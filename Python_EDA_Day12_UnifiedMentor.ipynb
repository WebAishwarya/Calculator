# 12.1 Pandas Duplicate Values

```python
import pandas as pd
import numpy as np
```

```python
# Duplicates

data = pd.DataFrame({'team': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B'],
                   'points': [10, 10, 12, 12, 15, 17, 20, 20],
                   'assists': [5, 5, 7, 9, 12, 9, 6, 6]})
```

```python
data
```

```python
data[data.duplicated()]
```

```python
team = data.duplicated(['team'])
data[team]
```

```python
team_pts = data.duplicated(['team','points'])
data[team_pts]
```

```python
data.drop_duplicates()
```

```python
data.drop_duplicates(['team'])
```

```python
data.drop_duplicates(['team','points'])
```

# 12.2 Handling Outliers Using IQR

```python
values = [1,2,3,4,5,6,9,10,12,55,60]
```

```python
values
```

```python
Q1 = np.percentile(values,25)
```

```python
Q1
```

```python
Q3 = np.percentile(values,75)
```

```python
Q3
```

```python
IQR = Q3 - Q1
```

```python
IQR
```

```python
low = Q1 - 1.5 * IQR      # 1.5 is constant
upp = Q3 + 1.5 * IQR
```

```python
low, upp
```

```python
Outliers = []
```

```python
values
```

```python
for i in values:
  if ((i > upp) or (i < low)):
    Outliers.append(i)
```

```python
Outliers
```

```python
for i in values:
  if i not in Outliers:
    print(i)
```

```python
values = [i for i in values if i not in Outliers]
```

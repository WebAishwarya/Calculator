# 9.2 Groupby & Date Range

# Groupby Function

```python
data2 = pd.DataFrame({
    'Ball' : [1,2,3,4,5],
    'X' : [270, 254, 142, 208, 158],
    'Y' : [163, 76, 156, 155, 343],
    'Outcome' : ['Dots', 'Dots', 'Dots', 'Dots', 'Dots'],
    'Phase' : ['Power Play', 'Death Overs', 'Middle Overs', 'Power Play', 'Death Overs']
})
```

```python
data2.to_csv('cricket.csv', index = False)
```

```python
cricket = pd.read_csv('/content/cricket.csv')
```

```python
cricket.head()
```

```python
cricket.groupby('Phase')[['Ball']].count()
```

```python
cricket.groupby(['Phase','Outcome'])[['Ball']].count()
```

# Pandas Date Range

```python
# Year-Month-Date   - Days Frequency

pd.date_range(start = '2024-01-01', end = '2024-12-31')
```

```python
# Months Frequency

pd.date_range(start = '2024-01-01', end = '2024-05-11', freq = 'M')
```

```python
# Weekdays Frequency

pd.date_range(start = '2024-01-01', end = '2024-01-10', freq = 'B')
```

# 9.3 Pandas Top Commands

Creating New, Renaming & Droping Columns.

Apply Method, Sort Index, Sort Values.

```python
data
```

**Creating New Columns**

```python
# Creating New Columns

data['City'] = ['Delhi', 'Bangalore', 'Chennai', 'Patna', 'Pune', 'Pune', 'Mumbai']
```

```python
data
```

**Renaming Columns**

```python
data.columns = ['NAME','AGE','UNIVERSITY','CITY']
```

```python
data
```

**Droping Columns**

```python
data.drop(['AGE'], axis = 1)           # Temporary Droping Syntax
```

```python
data
```

```python
data.drop(['AGE'], axis = 1, inplace = True)           # Permanent Droping Syntax
```

```python
data
```

**Apply Method**

```python
# Creating Marks Column

data['MARKS'] = [201, 210, 195, 183, 225, 203, 190]
```

```python
data
```

```python
# Apply Method

data['MARKS'] = data['MARKS'].apply(lambda x: x/5)
```

```python
data
```

```python
# Renaming Column

data.columns = ['NAME','UNIVERSITY','CITY','MARKS/SUBJECT']
```

```python
data
```

**Sort Index**

```python
data.index = [10,28,64,7,1,6,15]          # Assigning Index
```

```python
data
```

```python
data.sort_index()                      # Default Statement : Ascending = True
```

```python
data.sort_index(ascending = False)
```

**Sort Values**

```python
data.sort_values(by = 'MARKS/SUBJECT')      # Default Statement : Ascending = True
```

```python
data.sort_values(by = 'MARKS/SUBJECT', ascending = False)
```
